```python
# importing liberaries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# checking for the datasets in seaborn
sns.get_dataset_names()
```

```python
# chosing the dataset
df= sns.load_dataset('planets')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# checking for percentage of null values in the dataset
df.isnull().sum()/len(df)*100
```

```python
# checking null values in orbital_period column by visualization

sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
```

## As we can see that three columns have missing values including: orbital_period, mass, distance

```python
# removing missing values from orbital_period column and replacing it by its median
df['orbital_period'].fillna(df['orbital_period'].median(), inplace=True)
df['orbital_period'].isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
```

```python
df['mass'].fillna(df['orbital_period'].median(), inplace=True)
df['distance'].fillna(df['orbital_period'].median(), inplace=True)
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
```

we can see in the plot above that every column containing null values is filled either by it mean , median and mode.
